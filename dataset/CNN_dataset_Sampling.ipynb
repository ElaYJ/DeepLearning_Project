# Random Sampling

## train

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/DLP/train_meta_data.csv")
df
```

```python
tp_list =  df.query("growth_stage=='정식기'")['fname'].values.tolist()
fb_list = df.query("growth_stage=='출뢰기'")['fname'].values.tolist()
bs_list = df.query("growth_stage=='개화기'")['fname'].values.tolist()
fe_list = df.query("growth_stage=='과실비대기'")['fname'].values.tolist()
hv_list = df.query("growth_stage=='수확기'")['fname'].values.tolist()
len(tp_list), len(fb_list), len(bs_list), len(fe_list), len(hv_list)
```

```python
import random

t_transplant = random.sample(population=tp_list, k=int(len(tp_list)*0.45))
t_flowerbud = random.sample(population=fb_list, k=int(len(fb_list)*0.35))
t_blossoming = random.sample(population=bs_list, k=len(bs_list))
t_fruitenlarge = random.sample(population=fe_list, k=int(len(fe_list)*0.45))
t_harvesting = random.sample(population=hv_list, k=int(len(hv_list)*0.3))
len(t_transplant), len(t_flowerbud), len(t_blossoming), len(t_fruitenlarge), len(t_harvesting)
```

```python
len(t_transplant)+len(t_flowerbud)+len(t_blossoming)+len(t_fruitenlarge)+len(t_harvesting)
```

## valid & test

```python
vdf = pd.read_csv("/content/drive/MyDrive/DLP/valid_meta_data.csv")
vdf
```

```python
tp_list = vdf.query("growth_stage=='정식기'")['fname'].values.tolist()
fb_list = vdf.query("growth_stage=='출뢰기'")['fname'].values.tolist()
bs_list = df.query("growth_stage=='개화기'")['fname'].values.tolist()
fe_list = vdf.query("growth_stage=='과실비대기'")['fname'].values.tolist()
hv_list = vdf.query("growth_stage=='수확기'")['fname'].values.tolist()
len(tp_list), len(fb_list), len(bs_list), len(fe_list), len(hv_list)
```

```python
v_transplant = random.sample(population=tp_list, k=int(len(tp_list)*0.5))
v_flowerbud = random.sample(population=fb_list, k=int(len(fb_list)*0.3))
v_blossoming = random.sample(population=bs_list, k=200)
v_fruitenlarge = random.sample(population=fe_list, k=int(len(fe_list)*0.5))
v_harvesting = random.sample(population=hv_list, k=int(len(hv_list)*0.25))
len(v_transplant), len(v_flowerbud), len(v_blossoming), len(v_fruitenlarge), len(v_harvesting)
```

```python
len(v_transplant)+len(v_flowerbud)+len(v_blossoming)+len(v_fruitenlarge)+len(v_harvesting)
```

```python
ts_transplant = [x for x in tp_list if x not in v_transplant]
ts_flowerbud = random.sample(population=[x for x in fb_list if x not in v_flowerbud], k=int(len(fb_list)*0.3))
ts_blossoming = vdf.query("growth_stage=='개화기'")['fname'].values.tolist()
ts_fruitenlarge = [x for x in fe_list if x not in v_fruitenlarge]
ts_harvesting = random.sample(population=[x for x in hv_list if x not in v_harvesting], k=int(len(hv_list)*0.25))
len(ts_transplant), len(ts_flowerbud), len(ts_blossoming), len(ts_fruitenlarge), len(ts_harvesting)
```

```python
len(ts_transplant)+len(ts_flowerbud)+len(ts_blossoming)+len(ts_fruitenlarge)+len(ts_harvesting)
```

# to DataFrame

```python
train_fnames = dict(
    transplant = t_transplant,
    flowerbud = t_flowerbud,
    blossoming = t_blossoming,
    fruitenlarge = t_fruitenlarge,
    harvesting = t_harvesting
)
```

```python
valid_fnames = dict(
    transplant = v_transplant,
    flowerbud = v_flowerbud,
    blossoming = v_blossoming,
    fruitenlarge = v_fruitenlarge,
    harvesting = v_harvesting
)
```

```python
test_fnames = dict(
    transplant = ts_transplant,
    flowerbud = ts_flowerbud,
    blossoming = ts_blossoming,
    fruitenlarge = ts_fruitenlarge,
    harvesting = ts_harvesting
)
```

```python
import os

cls_list = ["transplant","flowerbud","blossoming","fruitenlarge","harvesting"]
tsrc_dir = "/content/drive/MyDrive/DLP/[DS]_Deep_Learning_Project/strawberry/train/images/"
vsrc_dir = "/content/drive/MyDrive/DLP/[DS]_Deep_Learning_Project/strawberry/valid/images/"

train_ds = {'fpath':[], 'class':[]}
valid_ds = {'fpath':[], 'class':[]}
test_ds = {'fpath':[], 'class':[]}
for i, cls in enumerate(cls_list):
    print("Train size(",cls,")", len(train_fnames[cls]))
    for fname in train_fnames[cls]:
        train_ds['fpath'].append(os.path.join(tsrc_dir, fname+".jpg"))
        train_ds['class'].append(i)

    print("Valid size(",cls,")", len(valid_fnames[cls]))
    for fname in valid_fnames[cls]:
        if cls == 'blossoming':
            valid_ds['fpath'].append(os.path.join(tsrc_dir, fname+".jpg"))
        else:
            valid_ds['fpath'].append(os.path.join(vsrc_dir, fname+".jpg"))
        valid_ds['class'].append(i)

    print("Test size(",cls,")", len(test_fnames[cls]))
    for fname in test_fnames[cls]:
        test_ds['fpath'].append(os.path.join(vsrc_dir, fname+".jpg"))
        test_ds['class'].append(i)
```

```python
train_df = pd.DataFrame(train_ds)
train_df = train_df.sample(frac=1)
train_df
```

```python
train_df.to_csv("/content/drive/MyDrive/DLP/growth_train2.csv", index=False)
```

```python
valid_df = pd.DataFrame(valid_ds)
valid_df = valid_df.sample(frac=1)
valid_df
```

```python
valid_df.to_csv("/content/drive/MyDrive/DLP/growth_valid2.csv", index=False)
```

```python
test_df = pd.DataFrame(test_ds)
test_df = test_df.sample(frac=1)
test_df
```

```python
test_df.to_csv("/content/drive/MyDrive/DLP/growth_test2.csv", index=False)
```

# EDA

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/DLP/growth_train2.csv")
df.head(2)
```

```python
cls_dict={0:"transplant",1:"flowerbud",2:"blossoming",3:"fruitenlarge",4:"harvesting"}
cls_dict.values()
```

```python
df['class'] = df['class'].map(cls_dict)
df.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# 축 서식 지정
def format_spines(ax, right_ax=False):
    # Setting up colors
    ax.spines['bottom'].set_color('#CCCCCC') #--> 라이트 그레이(Light Gray)
    ax.spines['left'].set_color('#CCCCCC')
    ax.spines['top'].set_visible(False) #--> top 축은 그리지 않는다.
    if right_ax:
        ax.spines['right'].set_color('#CCCCCC')
    else:
        ax.spines['right'].set_color('#FFFFFF') #--> 흰색
    ax.patch.set_facecolor('#FFFFFF')


def set_axis(ax, title='', xlabel='', ylabel='', angle=None, pad=20, right_ax=False):
    format_spines(ax, right_ax)
    if angle:
        for tick in ax.get_xticklabels():
            tick.set(fontsize=12)
            tick.set_rotation(angle)
    else:
        for tick in ax.get_xticklabels():
            tick.set(fontsize=12)
    if len(xlabel):   ax.set_xlabel(xlabel, fontsize=13)
    if len(ylabel):   ax.set_ylabel(ylabel, fontsize=13)
    if len(title):    ax.set_title(title, size=17, weight="bold", color='dimgrey', pad=pad)


def text_annotate(ax, with_p=False, ncount=None):
    for p in ax.patches:
        if with_p:
            x = p.get_bbox().get_points()[:, 0]
            y = p.get_bbox().get_points()[1, 1]
            ax.annotate(
				text='{}\n{:.1f}%'.format(int(y), 100. * y / ncount), xy=(x.mean(), y),
				ha='center', va='bottom', fontsize=12, color='darkblue'
			)
        else:
            ax.annotate(
				text=f"{int(p.get_height())}", xy=(p.get_x()+p.get_width()/2, p.get_height()), xytext=(0,7),
				ha='center', va='center', textcoords='offset points', fontsize=12, color='darkblue'
			)
```

```python
fig = plt.figure(constrained_layout=True, figsize=(7, 5))

# GridSpec
gs = GridSpec(1,1, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])

# ax1
sns.countplot(
    data=df, x='class', ax=ax1, hue='class', legend=False, palette='Dark2',
    order=list(cls_dict.values())
)
ncount = len(df)
text_annotate(ax=ax1, with_p=True, ncount=ncount)
set_axis(ax=ax1, title="Count per Growth Stage", xlabel="growth_stage", ylabel="count", pad=30)
```
