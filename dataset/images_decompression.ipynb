```python
from google.colab import drive
import os
import shutil
import tarfile
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## 이미지 파일 정리

```python
drive.mount('/content/drive')
```

#### Train

```python
# 디렉토리 경로
dir = '/content/drive/MyDrive/Colab Notebooks/Train/원천데이터'

# 디렉토리 파일 목록 가져오기
file_list = os.listdir(dir)

# tar 파일 선택
tar_files = [f for f in file_list if f.endswith('.tar')]

# tar 파일 압축 해제
for tar_file in tar_files:
    tar_file_path = os.path.join(dir, tar_file)
    with tarfile.open(tar_file_path, 'r') as tar:
        tar.extractall(path=dir)

print("압축 해제 완료")
```

```python
# 이동할 디렉토리
dir_end = '/content/drive/MyDrive/[DS]_Deep_Learning_Project/strawberry/train/images'

# dir에서 하위 디렉토리 목록을 가져옴
subdirs = [f for f in os.listdir(dir) if os.path.isdir(os.path.join(dir, f))]

# 각 하위 디렉토리의 파일들을 dir_end로 이동
for subdir in subdirs:
    subdir_path = os.path.join(dir, subdir)
    files = os.listdir(subdir_path)
    for file in files:
        shutil.move(os.path.join(subdir_path, file), dir_end)
```

#### valid

```python
# 디렉토리 경로
dir = '/content/drive/MyDrive/Colab Notebooks/Valid/원천데이터'

# 디렉토리 파일 목록 가져오기
file_list = os.listdir(dir)

# tar 파일 선택
tar_files = [f for f in file_list if f.endswith('.tar')]

# tar 파일 압축 해제
for tar_file in tar_files:
    tar_file_path = os.path.join(dir, tar_file)
    with tarfile.open(tar_file_path, 'r') as tar:
        tar.extractall(path=dir)

print("압축 해제 완료")
```
